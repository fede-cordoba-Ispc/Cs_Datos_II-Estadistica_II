###NOTA ACLARATORIA
La columna Academic_Performance: Este sistema de puntuación es común en las escuelas secundarias (High School) de Estados Unidos, se expresa con una escala numérica y su correspondiente letra. La escala de letras va de A (excelente) a F (reprobado). Esta calificación se usa para calcular el GPA (Grade Point Average), que es un promedio ponderado de todas las notas del estudiante y un factor clave para la admisión a la universidad.  
Aproximadamente, la correspondencia es la siguiente:  
A: 90 - 100 (Excelente)  
B: 80 - 89 (Bueno)  
C: 70 - 79 (Aceptable)  
D: 60 - 69 (Deficiente)  
En Argentina, el sistema de calificación más utilizado es una escala numérica del 1 al 10. No existe un sistema unificado para todo el país, y la nota mínima para aprobar puede variar entre un 6 y un 7 dependiendo de la provincia y la institución educativa.  
La comparación aproximada sería la siguiente:  
Calificación en EE.UU. (Escala de 100)  
Calificación en Argentina (Escala de 10)  
90-100 -> 9-10 (Sobresaliente)  
80-89 -> 7-8 (Notable)  
70-79 -> 6-7 (Aprobado/Bueno)  
60-69 -> 4-5 (Insuficiente/Desaprobado)  
< 60  -> 1-3 (Reprobado)

Es importante destacar que el sistema educativo estadounidense valora el trabajo diario, los proyectos y las participaciones, mientras que el argentino, aunque ha incorporado más evaluaciones continuas, tradicionalmente se ha enfocado en las notas de los exámenes.

**Qué hace este código**



1.   Detecta las columnas que son de tipo object o category (categóricas).
2.   Para cada columna, muestra los valores únicos y cuántas veces aparecen, incluyendo los valores faltantes (NaN).


3.   Esto te permite ver inconsistencias, como:

  *   Diferencias de mayúsculas/minúsculas (Sí vs si)

  *   Errores de tipeo (Male vs male)

  *   Valores inesperados o nulos.




````python
# 1. Cargar dataset en df
url = "https://raw.githubusercontent.com/fede-cordoba-Ispc/Cs_Datos_II-Estadistica_II/refs/heads/main/Datos/teen_phone_addiction_dataset.csv"
df = pd.read_csv(url)

# 2. Detectar columnas categóricas
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Columnas categóricas:", categorical_cols)

# 3. Función para detectar inconsistencias comunes
def analizar_categoricas(df, cols):
    for col in cols:
        print(f"\n--- Columna: {col} ---")
        # Contar valores únicos
        counts = df[col].value_counts(dropna=False)
        print(counts)

        # Detectar posibles inconsistencias
        inconsistencias = []
        for val in counts.index:
            if isinstance(val, str):
                # Verificar mayúsculas/minúsculas
                lower_vals = [v for v in counts.index if isinstance(v, str) and v.lower() == val.lower()]
                if len(lower_vals) > 1:
                    inconsistencias.append(f"Diferencias de mayúsculas/minúsculas: {lower_vals}")

                # Detectar espacios extra
                if val != val.strip():
                    inconsistencias.append(f"Espacios al inicio/final: '{val}'")

        if inconsistencias:
            print("Posibles inconsistencias detectadas:")
            for inc in inconsistencias:
                print("   -", inc)
        else:
            print("Sin inconsistencias obvias")

# 4. Ejecutar análisis
analizar_categoricas(df, categorical_cols)
````

# 3. Proponer y Manejar Outliers
Usamos el método más común y robusto para manejar outliers es el Rango Intercuartílico (IQR). La idea es identificar y, si es necesario, eliminar los valores que están significativamente alejados del centro de los datos, generalmente más allá de 1.5 veces el IQR de los cuartiles. Es crucial analizar cada caso, ya que a veces los outliers son datos válidos.

````python
# Definir la columna para el análisis de outliers
columna = 'Daily_Usage_Hours'

# Calcular el primer cuartil (Q1) y el tercer cuartil (Q3)
Q1 = df[columna].quantile(0.25)
Q3 = df[columna].quantile(0.75)

# Calcular el Rango Intercuartílico (IQR)
IQR = Q3 - Q1

# Definir los límites para los outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar y contar los outliers
outliers = df[(df[columna] < lower_bound) | (df[columna] > upper_bound)]

print(f" Se han detectado {len(outliers)} outliers en la columna '{columna}'.")
print(f"Esto representa el {round(len(outliers)/len(df)*100,2)}% del total de las filas.")
````

Explicación del Método
* Cálculo de Cuartiles y IQR: Primero, se calculan el primer cuartil (Q1) y el tercer cuartil (Q3). El Q1 representa el 25% de los datos y el Q3 el 75%. El Rango Intercuartílico (IQR) es simplemente la diferencia entre el Q3 y el Q1, y abarca el 50% central de los datos.

* Definición de Límites: Los límites para los outliers se definen como aquellos valores que están más allá de 1.5×IQR por encima del Q3 o por debajo del Q1. Los valores que caen fuera de estos límites se consideran outliers.

**lower_bound = Q1 - 1.5 * IQR**

**upper_bound = Q3 + 1.5 * IQR**

Interpretación de los Resultados

El código finaliza con dos líneas que imprimen el número de outliers detectados y su porcentaje con respecto al total de datos.

Se han detectado {len(outliers)} outliers en la columna 'Daily_Usage_Hours'.: Esta línea te dirá exactamente cuántos registros tienen un número de horas de uso diario que se considera atípico según el método del IQR.

Esto representa el {round(len(outliers)/len(df)*100,2)}% del total de las filas.: Esta métrica es crucial para entender la magnitud del problema. Por ejemplo, si el porcentaje es muy bajo (ej. menos del 1%), los outliers pueden ser solo casos aislados que no afectan demasiado el análisis. Sin embargo, si el porcentaje es alto, podría indicar una distribución de datos sesgada o la necesidad de una investigación más profunda.

###**ANOVA**


Se propone realizar un Análisis de Varianza (ANOVA) multifactorial, una técnica estadística que nos permite examinar cómo el promedio de una variable numérica (la variable dependiente) se ve afectado por dos o más variables categóricas (las variables independientes).

**Nuestro objetivo es doble:**
*   Evaluar el efecto de cada variable categórica por separado (efectos principales) en la variable dependiente.
*   Determinar si el efecto de una variable categórica cambia dependiendo del nivel de otra variable categórica (efectos de interacción).

A continuación, aplicaremos este análisis para explorar dos temas clave. Primero, investigaremos los factores que influyen en el Nivel de Adicción percibido. Luego, repetiremos el proceso para entender mejor los determinantes del Rendimiento Académico

````python
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

#Identifica la variable dependiente (numérica) y las independientes (categóricas)
variable_numerica = 'Addiction_Level'
variable_categorica1 = 'School_Grade'
variable_categorica2 = 'Gender'
variable_categorica3 = 'Phone_Usage_Purpose'

# --- 1. Tabla de ANOVA ---
# 'C(variable_categorica)' le dice a statsmodels que la variable es categórica

modelo_anova1 = ols(f'{variable_numerica} ~ C({variable_categorica1}) * C({variable_categorica2}) * C({variable_categorica3})', data=df).fit()
# Imprimir el resumen de la tabla ANOVA
tabla_anova1 = sm.stats.anova_lm(modelo_anova1, typ=2)
print("### Resultados de la Tabla ANOVA sobre Nivel de Adicción Autoprecibido ###")
print(tabla_anova1)

# Interpretación rápida del p-valor
# Para una multi-way ANOVA, es necesario interpretar cada p-value en cada iteración.
print("\nInterpretación de los resultados de ANOVA:")
for index, row in tabla_anova1.iterrows():
    p_valor = row['PR(>F)']
    if p_valor < 0.05:
        print(f"- {index}: El p-valor ({p_valor:.3f}) es menor que 0.05. Existen diferencias significativas.")
    else:
        print(f"- {index}: El p-valor ({p_valor:.3f}) no es significativo. No hay una diferencia significativa.")


 #Identifica la variable dependiente (numérica) y la independiente (categórica)
variable_numerica = 'Academic_Performance'
variable_categorica1 = 'School_Grade'
variable_categorica2 = 'Gender'
variable_categorica3 = 'Phone_Usage_Purpose'

# --- 1. Tabla de ANOVA ---
# 'C(variable_categorica)' le dice a statsmodels que la variable es categórica

modelo_anova2 = ols(f'{variable_numerica} ~ C({variable_categorica1}) * C({variable_categorica2}) * C({variable_categorica3})', data=df).fit()
# Imprimir el resumen de la tabla ANOVA
tabla_anova2 = sm.stats.anova_lm(modelo_anova2, typ=2)
print("\n### Resultados de la Tabla ANOVA sobre Rendimiento Académico ###")
print(tabla_anova2)

# Interpretación rápida del p-valor
# Para una multi-way ANOVA, es necesario interpretar cada p-value en cada iteración.
print("\nInterpretación de los resultados de ANOVA:")
for index, row in tabla_anova2.iterrows():
    p_valor = row['PR(>F)']
    if p_valor < 0.05:
        print(f"- {index}: El p-valor ({p_valor:.3f}) es menor que 0.05. Existen diferencias significativas.")
    else:
        print(f"- {index}: El p-valor ({p_valor:.3f}) no es significativo. No hay una diferencia significativa.")

# Configura la figura para tener 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Gráfico 1: Interacción entre School_Grade y Gender
sns.pointplot(data=df, x=variable_categorica1, y='Addiction_Level',
              hue=variable_categorica2, errorbar=('ci', 95), ax=axs[0])
axs[0].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica2} (Addiction Level)')
axs[0].set_ylabel(f'Nivel Promedio de Addiction Level')
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Gráfico 2: Interacción entre School_Grade y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica1, y='Addiction_Level',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[1])
axs[1].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica3} (Addiction Level)')
axs[1].set_ylabel('') # No se necesita doble etiqueta en el eje Y
axs[1].grid(axis='y', linestyle='--', alpha=0.6)
axs[1].tick_params(axis='x', rotation=45) # Rota etiquetas si son largas

# Gráfico 3: Interacción entre Gender y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica2, y='Addiction_Level',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[2])
axs[2].set_title(f'Interacción: {variable_categorica2} vs {variable_categorica3} (Addiction Level)')
axs[2].set_ylabel('')
axs[2].grid(axis='y', linestyle='--', alpha=0.6)
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Configura la figura para tener 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Gráfico 1: Interacción entre School_Grade y Gender
sns.pointplot(data=df, x=variable_categorica1, y='Academic_Performance',
              hue=variable_categorica2, errorbar=('ci', 95), ax=axs[0])
axs[0].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica2} (Academic Performance)')
axs[0].set_ylabel(f'Nivel Promedio de Academic Performance')
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Gráfico 2: Interacción entre School_Grade y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica1, y='Academic_Performance',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[1])
axs[1].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica3} (Academic Performance)')
axs[1].set_ylabel('') # No se necesita doble etiqueta en el eje Y
axs[1].grid(axis='y', linestyle='--', alpha=0.6)
axs[1].tick_params(axis='x', rotation=45) # Rota etiquetas si son largas

# Gráfico 3: Interacción entre Gender y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica2, y='Academic_Performance',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[2])
axs[2].set_title(f'Interacción: {variable_categorica2} vs {variable_categorica3} (Academic Performance)')
axs[2].set_ylabel('')
axs[2].grid(axis='y', linestyle='--', alpha=0.6)
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
````

````python
#  Definir el conjunto a analizar para VIF (el 80% de entrenamiento)
X_vif_train = X_train.copy()

#  Inicializar el DataFrame para guardar los resultados del VIF
vif_data_train = pd.DataFrame()
vif_data_train["feature"] = X_vif_train.columns

#  Calcular el VIF para cada variable en X_train
# Se itera sobre las columnas de X_train para calcular el VIF
vif_data_train["VIF"] = [variance_inflation_factor(X_vif_train.values, i)
                        for i in range(len(X_vif_train.columns))]

print("\n--- VIF del Conjunto de Entrenamiento (80%) ---")
print(vif_data_train)
````

 Un VIF que fue aproximadamente 4.38, similar al VIF del segudo modelo. Esto indica una multicolinealidad moderada, pero aceptable.

Ahora intentaremos inlcuir la constante (intercepto) es una mejora necesaria para evaluar la multicolinealidad de una manera que refleje fielmente el modelo de regresión final donde sí esta incluida.

Se agrega la constante usando sm.add_constant() porque queremos asegurarnos de que el modelo completo, incluyendo el intercepto, no sufra de inestabilidad causada por la multicolinealidad.

Se recomienda centrar usando la media del conjunto de entrenamiento (X_train)
para evitar fugas de datos (data leakage) del conjunto de prueba.

````python

mean_usage = X_train['Daily_Usage_Hours'].mean()
mean_apps = X_train['Apps_Used_Daily'].mean()


# Creamos una copia del conjunto de entrenamiento para no modificar el original
X_train_centered = X_train.copy()

# Centramos cada columna: restamos la media a cada valor
X_train_centered['Daily_Usage_Hours'] = X_train_centered['Daily_Usage_Hours'] - mean_usage
X_train_centered['Apps_Used_Daily'] = X_train_centered['Apps_Used_Daily'] - mean_apps

# Añadimos la constante (columna de unos) al DataFrame centrado
X_vif_train_final = sm.add_constant(X_train_centered, prepend=False)

# Inicializar el DataFrame para guardar los resultados del VIF
vif_data_final = pd.DataFrame()
vif_data_final["feature"] = X_vif_train_final.columns

vif_data_final["VIF"] = [variance_inflation_factor(X_vif_train_final.values, i)
                        for i in range(len(X_vif_train_final.columns))]


print("\n--- VIF del Conjunto de Entrenamiento (Centrado) ---")
print(vif_data_final)
````

## Resultados y Justificación del Centrado de Variables

Tras realizar la transformación de **Centrado de Variables** en el conjunto de entrenamiento (*X_train*), se recalculó el Factor de Inflación de Varianza (VIF), **incluyendo la constante** (intercepto), con los siguiente resultados mostrados
### 1. Análisis de Multicolinealidad Resuelta

El análisis final del VIF confirma que el centrado de las variables predictoras ha resuelto el problema de inestabilidad detectado en el intercepto, logrando una condición óptima para el modelo de regresión:
aprox 1.00:  
**Los valores de VIF para**  *Daily_Usage_Hours* y *Apps_Used_Daily* se mantienen extremadamente bajos y cercanos a $1$. Esto confirma que **no existe multicolinealidad** entre las dos variables predictoras en el conjunto de entrenamiento.


### 2. Implicación de la Transformación (Centrado)

El **centrado** es una transformación de escala estándar aplicada por las siguientes razones:

1.  **Estabilidad del Intercepto:** El alto VIF de la constante indicaba que el modelo tendría dificultades para estimar el intercepto ($\beta_0$) de manera fiable, dado que los valores de las variables $X$ estaban lejos de cero. Al restar la media, la columna de la constante deja de ser redundante con las variables $X$, logrando que su VIF caiga a $1$.
2.  **No Manipulación:** Es fundamental señalar que esta transformación **no altera la correlación intrínseca** entre *Daily_Usage_Hours* y *Apps_Used_Daily*  (de ahí que sus VIFs se mantengan bajos). Simplemente reajusta el punto de origen del modelo para mejorar la estabilidad de los coeficientes de regresión.

**Conclusión:** Con todos los VIF por debajo del umbral de $5$, las variables predictoras (*Daily_Usage_Hours* y *Apps_Used_Daily*) se consideran **altamente estables** y la matriz de diseño está en la mejor condición posible para la estimación del modelo de Regresión Lineal Múltiple.

# **El modelo final**

Finalmente, se usan los datos de entrenamiento mejorados para correr el modelo nuevamente.

````python

X_modelo = X_vif_train_final

modelo_ols = sm.OLS(y_train, X_modelo)

#  Entrenar el modelo
resultados = modelo_ols.fit()

print("\n--- RESUMEN DE LA REGRESIÓN LINEAL MULTIPLE (Entrenamiento Centrado) ---")
print(resultados.summary())
````

*Apps_Used_Daily* contribuye con un aumento de 0.11 unidades por cada aplicación adicional.  

**Comparación y Mejora**

# Diagnóstico Visual de Supuestos:
El objetivo de este análisis visual es asegurar que los errores del modelo son aleatorios, constantes y normales, permitiendo que las conclusiones extraídas de los coeficientes sean robustas.

````python
# Obtener Residuos y Valores Ajustados del modelo
residuos = resultados.resid
valores_ajustados = resultados.fittedvalues

# ----------------------------------------------------------------------
# --- GRÁFICO 1: Residuos vs. Valores Ajustados (Homocedasticidad) ---
# ----------------------------------------------------------------------

plt.figure(figsize=(10, 6))
sns.scatterplot(x=valores_ajustados, y=residuos, alpha=0.6)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.xlabel("Valores Ajustados (Predicted Addiction Level)")
plt.ylabel("Residuos (Errores)")
plt.title("Gráfico 1: Residuos vs. Valores Ajustados")
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
````

**Fallo de los Supuestos (Heterocedasticidad)**


 El Gráfico de Residuos vs. Valores Ajustados muestra un claro patrón de embudo invertido, lo que indica Heterocedasticidad (la varianza del error no es constante).

  Esto significa que la precisión del modelo es baja para niveles de adicción bajos y aumenta para niveles altos.
  
 Violación del Rango de la Variable (Límites) El modelo predice niveles de adicción (Valores Ajustados) superiores a 10 (llegando a 12), lo cual es imposible dado que tu variable dependiente está limitada físicamente al rango de 0 a 10.
    

````python
# ----------------------------------------------------------------------
# --- GRÁFICO 2: Histograma de Residuos (Normalidad Visual) ---
# ----------------------------------------------------------------------

plt.figure(figsize=(10, 6))
sns.histplot(residuos, kde=True, bins=30, color='skyblue')
plt.title("Gráfico 2: Distribución (Normalidad) de los Residuos")
plt.xlabel("Residuos")
plt.ylabel("Frecuencia")
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
````

## Análisis de la Distribución de los Residuos

El **Histograma de Residuos** es fundamental para evaluar el supuesto de **Normalidad de los Errores** en nuestro modelo de Regresión Lineal Múltiple. Visualmente, se observa que la distribución de los errores **no se ajusta a una curva normal perfecta** (la curva de campana, representada por la línea azul).

La principal desviación es la **asimetría negativa (sesgo a la izquierda)**, lo cual se evidencia en una "cola" más larga y pesada hacia los valores negativos (errores entre $-2$ y $-5$). Esto sugiere que el modelo OLS tiende a **sobreestimar** el Nivel de Adicción con mayor frecuencia que a subestimarlo, especialmente en ciertos rangos.


````python
# ----------------------------------------------------------------------
# --- GRÁFICO 3: Q-Q Plot de Residuos (Normalidad Detallada) ---
# ----------------------------------------------------------------------

fig = sm.qqplot(residuos, line='s') # 's' ajusta la línea a la distribución de residuos
plt.title("Gráfico 3: Q-Q Plot de los Residuos")
plt.show()
````

El Gráfico Q-Q Plot de los Residuos es la prueba visual más rigurosa para evaluar el supuesto de Normalidad de los Errores del modelo OLS.

Análisis del Gráfico
Desviación Central: Los puntos se desvían de la línea roja (que representa una distribución normal perfecta), especialmente en los extremos.

Colas Pesadas: La curva de los residuos cae por debajo de la línea en la cola inferior (valores negativos) y se desvía por encima de la línea en la cola superior (valores positivos). Esto confirma la distribución sesgada observada en el histograma y sugiere que los errores reales del modelo tienen colas más pesadas (más valores atípicos o extremos) que una distribución normal.

Confirmación del Sesgo: La desviación confirma el sesgo a la izquierda (asimétrico negativo) previamente identificado.

A pesar de que el modelo es estable (VIF bajo) y tiene un buen poder predictivo en los datos de prueba (MAE de $0.94$), el incumplimiento sistemático de los supuestos y la violación del rango $0-10$ sugieren el abandono del $\text{OLS}$ en favor de un modelo diseñado para datos acotados.


# Ahora se probara al modelo con los datos de prueba (20)



````python
from sklearn.metrics import mean_absolute_error, mean_squared_error


X_test_centered = X_test.copy()

# Centramos CADA columna de prueba usando las MEDIAS del entrenamiento
X_test_centered['Daily_Usage_Hours'] = X_test_centered['Daily_Usage_Hours'] - mean_usage
X_test_centered['Apps_Used_Daily'] = X_test_centered['Apps_Used_Daily'] - mean_apps

# Esto crea el DataFrame final listo para la predicción
X_test_final = sm.add_constant(X_test_centered, prepend=False)

# Hacemos la predicción con los datos de prueba centrados
y_predicho = resultados.predict(X_test_final)

# Calcular el MAE (Error Absoluto Medio)
mae = mean_absolute_error(y_test, y_predicho)

# Calcular el RMSE (Raíz del Error Cuadrático Medio)
rmse = np.sqrt(mean_squared_error(y_test, y_predicho))

print("\n--- Evaluación del Rendimiento (Datos de Prueba 20%) ---")
print(f"Error Absoluto Medio (MAE): {mae:.4f}")
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse:.4f}")
````
