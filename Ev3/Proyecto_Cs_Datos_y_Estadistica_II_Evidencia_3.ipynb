###NOTA ACLARATORIA
La columna Academic_Performance: Este sistema de puntuación es común en las escuelas secundarias (High School) de Estados Unidos, se expresa con una escala numérica y su correspondiente letra. La escala de letras va de A (excelente) a F (reprobado). Esta calificación se usa para calcular el GPA (Grade Point Average), que es un promedio ponderado de todas las notas del estudiante y un factor clave para la admisión a la universidad.  
Aproximadamente, la correspondencia es la siguiente:  
A: 90 - 100 (Excelente)  
B: 80 - 89 (Bueno)  
C: 70 - 79 (Aceptable)  
D: 60 - 69 (Deficiente)  
En Argentina, el sistema de calificación más utilizado es una escala numérica del 1 al 10. No existe un sistema unificado para todo el país, y la nota mínima para aprobar puede variar entre un 6 y un 7 dependiendo de la provincia y la institución educativa.  
La comparación aproximada sería la siguiente:  
Calificación en EE.UU. (Escala de 100)  
Calificación en Argentina (Escala de 10)  
90-100 -> 9-10 (Sobresaliente)  
80-89 -> 7-8 (Notable)  
70-79 -> 6-7 (Aprobado/Bueno)  
60-69 -> 4-5 (Insuficiente/Desaprobado)  
< 60  -> 1-3 (Reprobado)

Es importante destacar que el sistema educativo estadounidense valora el trabajo diario, los proyectos y las participaciones, mientras que el argentino, aunque ha incorporado más evaluaciones continuas, tradicionalmente se ha enfocado en las notas de los exámenes.

**Qué hace este código**



1.   Detecta las columnas que son de tipo object o category (categóricas).
2.   Para cada columna, muestra los valores únicos y cuántas veces aparecen, incluyendo los valores faltantes (NaN).


3.   Esto te permite ver inconsistencias, como:

  *   Diferencias de mayúsculas/minúsculas (Sí vs si)

  *   Errores de tipeo (Male vs male)

  *   Valores inesperados o nulos.




````python
# 1. Cargar dataset en df
url = "https://raw.githubusercontent.com/fede-cordoba-Ispc/Cs_Datos_II-Estadistica_II/refs/heads/main/Datos/teen_phone_addiction_dataset.csv"
df = pd.read_csv(url)

# 2. Detectar columnas categóricas
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Columnas categóricas:", categorical_cols)

# 3. Función para detectar inconsistencias comunes
def analizar_categoricas(df, cols):
    for col in cols:
        print(f"\n--- Columna: {col} ---")
        # Contar valores únicos
        counts = df[col].value_counts(dropna=False)
        print(counts)

        # Detectar posibles inconsistencias
        inconsistencias = []
        for val in counts.index:
            if isinstance(val, str):
                # Verificar mayúsculas/minúsculas
                lower_vals = [v for v in counts.index if isinstance(v, str) and v.lower() == val.lower()]
                if len(lower_vals) > 1:
                    inconsistencias.append(f"Diferencias de mayúsculas/minúsculas: {lower_vals}")

                # Detectar espacios extra
                if val != val.strip():
                    inconsistencias.append(f"Espacios al inicio/final: '{val}'")

        if inconsistencias:
            print("Posibles inconsistencias detectadas:")
            for inc in inconsistencias:
                print("   -", inc)
        else:
            print("Sin inconsistencias obvias")

# 4. Ejecutar análisis
analizar_categoricas(df, categorical_cols)
````

# 3. Proponer y Manejar Outliers
Usamos el método más común y robusto para manejar outliers es el Rango Intercuartílico (IQR). La idea es identificar y, si es necesario, eliminar los valores que están significativamente alejados del centro de los datos, generalmente más allá de 1.5 veces el IQR de los cuartiles. Es crucial analizar cada caso, ya que a veces los outliers son datos válidos.

````python
# Definir la columna para el análisis de outliers
columna = 'Daily_Usage_Hours'

# Calcular el primer cuartil (Q1) y el tercer cuartil (Q3)
Q1 = df[columna].quantile(0.25)
Q3 = df[columna].quantile(0.75)

# Calcular el Rango Intercuartílico (IQR)
IQR = Q3 - Q1

# Definir los límites para los outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar y contar los outliers
outliers = df[(df[columna] < lower_bound) | (df[columna] > upper_bound)]

print(f" Se han detectado {len(outliers)} outliers en la columna '{columna}'.")
print(f"Esto representa el {round(len(outliers)/len(df)*100,2)}% del total de las filas.")
````

Explicación del Método
* Cálculo de Cuartiles y IQR: Primero, se calculan el primer cuartil (Q1) y el tercer cuartil (Q3). El Q1 representa el 25% de los datos y el Q3 el 75%. El Rango Intercuartílico (IQR) es simplemente la diferencia entre el Q3 y el Q1, y abarca el 50% central de los datos.

* Definición de Límites: Los límites para los outliers se definen como aquellos valores que están más allá de 1.5×IQR por encima del Q3 o por debajo del Q1. Los valores que caen fuera de estos límites se consideran outliers.

**lower_bound = Q1 - 1.5 * IQR**

**upper_bound = Q3 + 1.5 * IQR**

Interpretación de los Resultados

El código finaliza con dos líneas que imprimen el número de outliers detectados y su porcentaje con respecto al total de datos.

Se han detectado {len(outliers)} outliers en la columna 'Daily_Usage_Hours'.: Esta línea te dirá exactamente cuántos registros tienen un número de horas de uso diario que se considera atípico según el método del IQR.

Esto representa el {round(len(outliers)/len(df)*100,2)}% del total de las filas.: Esta métrica es crucial para entender la magnitud del problema. Por ejemplo, si el porcentaje es muy bajo (ej. menos del 1%), los outliers pueden ser solo casos aislados que no afectan demasiado el análisis. Sin embargo, si el porcentaje es alto, podría indicar una distribución de datos sesgada o la necesidad de una investigación más profunda.

###**ANOVA**


Se propone realizar un Análisis de Varianza (ANOVA) multifactorial, una técnica estadística que nos permite examinar cómo el promedio de una variable numérica (la variable dependiente) se ve afectado por dos o más variables categóricas (las variables independientes).

**Nuestro objetivo es doble:**
*   Evaluar el efecto de cada variable categórica por separado (efectos principales) en la variable dependiente.
*   Determinar si el efecto de una variable categórica cambia dependiendo del nivel de otra variable categórica (efectos de interacción).

A continuación, aplicaremos este análisis para explorar dos temas clave. Primero, investigaremos los factores que influyen en el Nivel de Adicción percibido. Luego, repetiremos el proceso para entender mejor los determinantes del Rendimiento Académico

````python
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

#Identifica la variable dependiente (numérica) y las independientes (categóricas)
variable_numerica = 'Addiction_Level'
variable_categorica1 = 'School_Grade'
variable_categorica2 = 'Gender'
variable_categorica3 = 'Phone_Usage_Purpose'

# --- 1. Tabla de ANOVA ---
# 'C(variable_categorica)' le dice a statsmodels que la variable es categórica

modelo_anova1 = ols(f'{variable_numerica} ~ C({variable_categorica1}) * C({variable_categorica2}) * C({variable_categorica3})', data=df).fit()
# Imprimir el resumen de la tabla ANOVA
tabla_anova1 = sm.stats.anova_lm(modelo_anova1, typ=2)
print("### Resultados de la Tabla ANOVA sobre Nivel de Adicción Autoprecibido ###")
print(tabla_anova1)

# Interpretación rápida del p-valor
# Para una multi-way ANOVA, es necesario interpretar cada p-value en cada iteración.
print("\nInterpretación de los resultados de ANOVA:")
for index, row in tabla_anova1.iterrows():
    p_valor = row['PR(>F)']
    if p_valor < 0.05:
        print(f"- {index}: El p-valor ({p_valor:.3f}) es menor que 0.05. Existen diferencias significativas.")
    else:
        print(f"- {index}: El p-valor ({p_valor:.3f}) no es significativo. No hay una diferencia significativa.")


 #Identifica la variable dependiente (numérica) y la independiente (categórica)
variable_numerica = 'Academic_Performance'
variable_categorica1 = 'School_Grade'
variable_categorica2 = 'Gender'
variable_categorica3 = 'Phone_Usage_Purpose'

# --- 1. Tabla de ANOVA ---
# 'C(variable_categorica)' le dice a statsmodels que la variable es categórica

modelo_anova2 = ols(f'{variable_numerica} ~ C({variable_categorica1}) * C({variable_categorica2}) * C({variable_categorica3})', data=df).fit()
# Imprimir el resumen de la tabla ANOVA
tabla_anova2 = sm.stats.anova_lm(modelo_anova2, typ=2)
print("\n### Resultados de la Tabla ANOVA sobre Rendimiento Académico ###")
print(tabla_anova2)

# Interpretación rápida del p-valor
# Para una multi-way ANOVA, es necesario interpretar cada p-value en cada iteración.
print("\nInterpretación de los resultados de ANOVA:")
for index, row in tabla_anova2.iterrows():
    p_valor = row['PR(>F)']
    if p_valor < 0.05:
        print(f"- {index}: El p-valor ({p_valor:.3f}) es menor que 0.05. Existen diferencias significativas.")
    else:
        print(f"- {index}: El p-valor ({p_valor:.3f}) no es significativo. No hay una diferencia significativa.")

# Configura la figura para tener 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Gráfico 1: Interacción entre School_Grade y Gender
sns.pointplot(data=df, x=variable_categorica1, y='Addiction_Level',
              hue=variable_categorica2, errorbar=('ci', 95), ax=axs[0])
axs[0].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica2} (Addiction Level)')
axs[0].set_ylabel(f'Nivel Promedio de Addiction Level')
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Gráfico 2: Interacción entre School_Grade y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica1, y='Addiction_Level',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[1])
axs[1].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica3} (Addiction Level)')
axs[1].set_ylabel('') # No se necesita doble etiqueta en el eje Y
axs[1].grid(axis='y', linestyle='--', alpha=0.6)
axs[1].tick_params(axis='x', rotation=45) # Rota etiquetas si son largas

# Gráfico 3: Interacción entre Gender y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica2, y='Addiction_Level',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[2])
axs[2].set_title(f'Interacción: {variable_categorica2} vs {variable_categorica3} (Addiction Level)')
axs[2].set_ylabel('')
axs[2].grid(axis='y', linestyle='--', alpha=0.6)
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Configura la figura para tener 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Gráfico 1: Interacción entre School_Grade y Gender
sns.pointplot(data=df, x=variable_categorica1, y='Academic_Performance',
              hue=variable_categorica2, errorbar=('ci', 95), ax=axs[0])
axs[0].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica2} (Academic Performance)')
axs[0].set_ylabel(f'Nivel Promedio de Academic Performance')
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Gráfico 2: Interacción entre School_Grade y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica1, y='Academic_Performance',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[1])
axs[1].set_title(f'Interacción: {variable_categorica1} vs {variable_categorica3} (Academic Performance)')
axs[1].set_ylabel('') # No se necesita doble etiqueta en el eje Y
axs[1].grid(axis='y', linestyle='--', alpha=0.6)
axs[1].tick_params(axis='x', rotation=45) # Rota etiquetas si son largas

# Gráfico 3: Interacción entre Gender y Phone_Usage_Purpose
sns.pointplot(data=df, x=variable_categorica2, y='Academic_Performance',
              hue=variable_categorica3, errorbar=('ci', 95), ax=axs[2])
axs[2].set_title(f'Interacción: {variable_categorica2} vs {variable_categorica3} (Academic Performance)')
axs[2].set_ylabel('')
axs[2].grid(axis='y', linestyle='--', alpha=0.6)
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
````

# Sugerencia del profesor trabajar  con  regresion lineal multiple  

Se eligen las variables independientes **(X)** a Daily_Usage_Hours, Apps_Used_Daily y Sleep_Hours y como variable independiente **(Y)** a Addiction_Level ya que son las que se observaron con mayor índice de correlación.

**Código Paso a Paso para la Regresión** **Múltiple**

````python
# Importar librerías necesarias

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor



# Definición de las variables para la Regresión Múltiple
Y = df['Addiction_Level']
X = df[['Daily_Usage_Hours', 'Apps_Used_Daily', 'Sleep_Hours']]

print("Variables del modelo múltiple definidas.")
````

**2. Detección de Multicolinealidad (VIF)**

Verificamos que no haya redundancia en los predictores.

````python
# Cálculo del VIF (Factor de Inflación de la Varianza)
X_no_const = X.copy()
vif_data = pd.DataFrame()
vif_data["feature"] = X_no_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_no_const.values, i)
                   for i in range(len(X_no_const.columns))]

print("\n--- Factor de Inflación de la Varianza (VIF) ---")
print(vif_data)

# Si todos los VIF son < 5, el modelo es fiable.
````

**Conclusiones:**

Los predictores muestan valores de VIF de moderado (6.22) a muy alto (8.99) todos superan el umbral de 5 acercándose uno de ellos al 10. ESto nos indica que hay redundancia de información y la variable Sleep_Hours es la que más contribuye a la multicolinealidad.  
Se decide ajustar el modelo sin contar con esta última variable.  


# Estrategia de Eliminación y Ajuste del Modelo


**1. Re-Definición de Variables**

````python
# NUEVO MODELO SIN Sleep_Hours
Y_nuevo = df['Addiction_Level']
X_nuevo = df[['Daily_Usage_Hours', 'Apps_Used_Daily']]
````

**2. Detección de Multicolinealidad (VIF) - ¡Segundo intento!**

````python
# Calcular el VIF para el nuevo conjunto reducido
X_vif_corr = X_nuevo.copy()
vif_data_corr = pd.DataFrame()
vif_data_corr["feature"] = X_vif_corr.columns
vif_data_corr["VIF"] = [variance_inflation_factor(X_vif_corr.values, i)
                        for i in range(len(X_vif_corr.columns))]

print("\n--- VIF del Modelo Corregido (Solo 2 Predictores) ---")
print(vif_data_corr)
````

**3. Ejecución del Modelo Múltiple Corregido**

````python
# 1. Agregar la constante (Intercepto)
X_con_constante_corr = sm.add_constant(X_nuevo)

# 2. Crear y ajustar el modelo OLS corregido
modelo_multiple_corregido = sm.OLS(Y_nuevo, X_con_constante_corr).fit()

# 3. Mostrar el resumen final del modelo
print("\n--- Resultados Finales de Regresión Lineal Múltiple (Corregido) ---")
print(modelo_multiple_corregido.summary())
````
